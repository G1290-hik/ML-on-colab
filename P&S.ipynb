```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
```

### The link to the database:
[www.kaggle.com/datasets/imbikramsaha/indian-gdp](https://)

```python
data=pd.read_csv('/content/India_GDP_Data.csv')
data.head()
```

```python
data.describe()
```

```python
plt.figure(figsize=(100, 6))  # Adjust width and height as needed
data.plot(kind='scatter', x='Year', y='GDP_In_Billion_USD', s=15, alpha=0.8)
plt.gca().spines[['top', 'right']].set_visible(False)
plt.xticks(data['Year'],rotation=90)  # Rotate x-axis labels for better readability
plt.gca().xaxis.set_major_locator(MultipleLocator(3))
plt.show()
```

## **Percentage Change in GDP Vs. GDP(in USD)**

```python
data.plot(kind='scatter', x='Per_Capita_in_USD', y='Percentage_Growth ', s=32, alpha=.8,color="red")
plt.gca().spines[['top', 'right',]].set_visible(False)
plt.xticks(data['Per_Capita_in_USD'],rotation=90)
plt.gca().xaxis.set_major_locator(MultipleLocator(100))
plt.show()
```

```python
def _plot_series(series, series_name, series_index=0):
    from matplotlib import pyplot as plt
    import seaborn as sns
    palette = list(sns.palettes.mpl_palette('Dark2'))
    xs = series['Year']
    ys = series['Percentage_Growth ']

    # Plot the series
    plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

    # Highlight the starting and ending points
    plt.scatter(xs.iloc[0], ys.iloc[0], color='green', label='Start', zorder=5)
    plt.scatter(xs.iloc[-1], ys.iloc[-1], color='red', label='End', zorder=5)

fig, ax = plt.subplots(figsize=(10, 7))
df_sorted = data.sort_values('Year', ascending=True)
_plot_series(df_sorted, 'Series')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Year')
plt.ylabel('Percentage Growth')
plt.xticks(rotation=90)
plt.gca().xaxis.set_major_locator(MultipleLocator(3))
plt.legend()
plt.show()
```
