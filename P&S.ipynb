```python
def format_number_in_scientific_notation(number):
    # Split the number into coefficient and exponent parts
    coefficient, exponent = "{:.4e}".format(number).split('e')

    # Convert the exponent to an integer
    exponent = int(exponent)

    # Format the coefficient part
    formatted_coefficient = f"{float(coefficient):.4f}"

    # Check if the exponent is smaller than the threshold
    if abs(exponent) < 3:
        return formatted_coefficient
    else:
        return f"{formatted_coefficient} x 10^{exponent}"
```

```python
def interpret_tail_of_distribution(statistic):
    if statistic > 0:
        return "The data shows a positively skewed tail."
    elif statistic < 0:
        return "The data shows a negatively skewed tail."
    else:
        return "The data shows a symmetric distribution."

def interpret_kurtosis(statistic):
    if statistic > 0:
        return "The data shows a heavier tail than a normal distribution (leptokurtic)."
    elif statistic < 0:
        return "The data shows a lighter tail than a normal distribution (platykurtic)."
    else:
        return "The data shows a tail similar to a normal distribution."
```

```python
#Calculate all the stats for the function
def stat_analysis(data, y_attribute):
    # Calculate mean, median, mode, kurtosis, and skewness for each attribute
    stats_dict = {}
    for column in data.columns:
        if column != y_attribute:
            mean = format_number_in_scientific_notation(data[column].mean())
            median = format_number_in_scientific_notation(data[column].median())
            mode = format_number_in_scientific_notation(data[column].mode().iloc[0])
            kurtosis = data[column].kurtosis()
            kurtosis_text = interpret_kurtosis(kurtosis)
            skewness = data[column].skew()
            skewness_text = interpret_tail_of_distribution(skewness)
            stats_dict[column] = {
                'Mean': mean,
                'Median': median,
                'Mode': mode,
                'Kurtosis': f"{kurtosis:.4f} - {kurtosis_text}",
                'Skewness': f"{skewness:.4f} - {skewness_text}"
            }

    return stats_dict
```

```python
stats = stat_analysis(data, 'Year')
print("Summary statistics for each attribute:")
for attribute, values in stats.items():
    print(f"\nAttribute: {attribute}")
    print("Mean:", values['Mean'])
    print("Median:", values['Median'])
    print("Mode:", values['Mode'])
    print("Kurtosis:", values['Kurtosis'])
    print("Skewness:", values['Skewness'])
```
